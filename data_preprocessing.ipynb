<h3> Add new column names "sales" and sales = price * product.
<h3> But direct multiplication is not possible as price has '$' sign and datatype is str.
<h3> So, need to convert to str 2st then replacing '$' with '' & then converting it to float.

```python
df_0['sales'] = df_0['price'].str.replace('$', '', regex=False).astype(float) * df_0['quantity']
```

<h3> dropping 'price' & 'quantity' cols

```python
df_0.drop(columns=['price', 'quantity'], inplace=True)
```

```python
df_0
```

<h3> saving the data in same csv file fpor "df_0".

```python
df_0.to_csv('.\\data\\processed_data_0.csv', index=False)
```

<h3> repeating same procedure for "df_1" & "df_2"

```python
df_1['sales'] = df_1['price'].str.replace('$', '', regex=False).astype(float)
df_2['sales'] = df_2['price'].str.replace('$', '', regex=False).astype(float)

df_1.drop(columns=['price', 'quantity'], inplace=True)
df_2.drop(columns=['price', 'quantity'], inplace=True)

df_1.to_csv('.\\data\\processed_data_1.csv', index=False)
df_2.to_csv('.\\data\\processed_data_2.csv', index=False)
```

<h3> now merging all 3 csv files

```python
merged = pd.concat([df_0, df_1, df_2], ignore_index=True)
merged.to_csv('.\\data\\merged_processed_data.csv', index=False)
```
