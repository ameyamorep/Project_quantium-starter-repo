```python
from dash import Dash, html, dcc
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv(r'C:\Users\think\Downloads\Ameya UTS Masters\Projects\Project_quantium-starter-repo\data\merged_processed_data.csv')
```

```python
app = Dash()
```

```python
fig = px.line(df, x = 'date', y = 'sales', title='Sales over time')

app.layout = html.Div(children=[html.H1(children='Sales Dashboard'), 
                                html.Div(children='''Dash: A web app for Sales Visualisation'''),
                                dcc.Graph(id='sales-graph', figure=fig)])
```

```python
if __name__ == '__main__':
    app.run(debug=True)
```
